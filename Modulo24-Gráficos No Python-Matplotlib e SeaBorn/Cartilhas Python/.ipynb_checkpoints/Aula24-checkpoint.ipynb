## ----------------------------Matplotlib-Primeiros exemplos------------------

```python
#Quando só se precisa de um tipo de gráfico na figura essa interface é chamada de pyplot 
#Aqui estão algumas funções pyplot comumente usadas:
#plt.figure(): Cria uma nova figura.
#plt.plot(): Plota y versus x como linhas e/ou marcadores.
#plt.scatter(): Plota um gráfico de dispersão de y versus x.
#plt.bar(): Plota um gráfico de barras.
#plt.xlabel(): Define o rótulo do eixo x.
#plt.ylabel(): Define o rótulo do eixo y.
#plt.title(): Define um título para os eixos.
#plt.grid(): Configura as linhas da grade.
#plt.legend(): Coloca uma legenda no eixo.
#plt.show(): Exibe uma figura.
```

```python
import matplotlib.pyplot as plt

meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]

plt.plot(meses,vendas)
plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt

meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]

plt.plot(meses,vendas, marker='o')
plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.grid(True)
plt.show()
```

```python
cidades=['São Paulo','Rio de Janeiro','Salvador']
temperaturas=[20,25,27]

plt.bar(cidades,temperaturas)
plt.xlabel('Cidades')
plt.ylabel('Temperatura Média (°C)')
plt.title('Comparação da Temperatura Média entre Cidades')
plt.show()
```

```python
cidades=['São Paulo','Rio de Janeiro','Salvador']
temperaturas=[20,25,27]

barras=plt.bar(cidades,temperaturas,color='green')

plt.bar_label(barras, labels=temperaturas, padding=3, fontsize=10, fontweight='bold')

plt.xlabel('Cidades')
plt.ylabel('Temperatura Média (°C)')
plt.title('Comparação da Temperatura Média entre Cidades')
plt.show()
```

## ----------------------------Paleta de cores parte 1------------------

```python
categorias=['Eletrônicos','Roupas','Alimentos','Livros']
quantidades=[45,30,15,10]
destaque=[0.1,0,0,0]#destacar a primeira fatia

plt.pie(quantidades,labels=categorias,autopct='%.1f', explode=destaque)
plt.title('Distribuição de Categorias de Produtos em Porcentagem')
plt.show()
```

```python
categorias=['Eletrônicos','Roupas','Alimentos','Livros']
quantidades=[45,30,15,10]
destaque=[0.1,0,0,0]#destacar a primeira fatia

cores=plt.get_cmap('Pastel2').colors

plt.pie(quantidades, labels=categorias , colors=cores , autopct='%.1f', explode=destaque)
plt.title('Distribuição de Categorias de Produtos em Porcentagem')
plt.show()
```

```python
anos=[2018,2019,2020,2021]
vendas_online=[5,10,15,20]
vendas_loja=[10,8,6,4]

plt.bar(anos,vendas_online,label='Online')
plt.bar(anos,vendas_loja,bottom=vendas_online,label='Loja')
plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show
```

```python
anos=[2018,2019,2020,2021]
vendas_online=[5,10,15,20]
vendas_loja=[10,8,6,4]

plt.bar(anos,vendas_online,label='Online')
plt.bar(anos,vendas_loja,bottom=vendas_online,label='Loja')

#Forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos,labels=anos)

plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show
```

```python
anos=[2018,2019,2020,2021]
vendas_online=[5,10,15,20]
vendas_loja=[10,8,6,4]

cores=plt.get_cmap('Pastel2').colors

plt.bar(anos,vendas_online,label='Online',color=cores[0])
plt.bar(anos,vendas_loja,bottom=vendas_online,label='Loja',color=cores[1])

#Forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos,labels=anos)

plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show
```

```python
import matplotlib.pyplot as plt
from cycler import cycler


anos=[2018,2019,2020,2021]
vendas_online=[5,10,15,20]
vendas_loja=[10,8,6,4]

cores_cycler=cycler('color',plt.get_cmap('Pastel2').colors)

#rc = run configuration(configuração quando executa)
plt.rc('axes',prop_cycle=cores_cycler)

plt.bar(anos,vendas_online,label='Online',color=cores[0])
plt.bar(anos,vendas_loja,bottom=vendas_online,label='Loja',color=cores[1])

#Forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos,labels=anos)

plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np

meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]

#Usando o np where, crie uma lista de marcadores x para o valor mínimo e o para o resto

marcadores=np.where(vendas==np.min(vendas),'x','')

plt.plot(meses,vendas)

for posicao,marcador in enumerate(marcadores):
    if marcador=='x':
        cor='red'
        tamanho=80
    else:
        cor='blue'
        tamanho=0
    plt.scatter(meses[posicao],vendas[posicao],marker=marcador,s=tamanho,color=cor)

#Adicionando uma anotação para mostrar o menor valor em março
plt.annotate('Menor Valor',color='red',xy=(2,78),xytext=(2.2,79))

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]
marcadores=np.where(vendas==np.min(vendas),'x','')

#Paleta de Cores
cores=plt.get_cmap('Dark2').colors
ciclo_cores=cycler('color',cores)
plt.rc('axes',prop_cycle=ciclo_cores)

#Gráfico Base
plt.plot(meses,vendas)

for posicao,marcador in enumerate(marcadores):
    if marcador=='x':
        cor='red'
        tamanho=80
    else:
        cor='blue'
        tamanho=0
    plt.scatter(meses[posicao],vendas[posicao],marker=marcador,s=tamanho,color=cor)

#Adicionando uma anotação para mostrar o menor valor em março
plt.annotate('Menor Valor',color='red',xy=(2,78),xytext=(2.2,79))

#Destaque de Região
plt.axvspan(6,11,alpha=0.2,color=cores[1])
plt.text(6,80,'Recuperação de Crescimento',color='green',fontsize=9,weight='bold')

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]
marcadores=np.where(vendas==np.min(vendas),'x','')

#Paleta de Cores
cores=plt.get_cmap('Dark2').colors
ciclo_cores=cycler('color',cores)
plt.rc('axes',prop_cycle=ciclo_cores)

#Gráfico Base
plt.plot(meses,vendas)

for posicao,marcador in enumerate(marcadores):
    if marcador=='x':
        cor='red'
        tamanho=80
    else:
        cor='limegreen'
        tamanho=0
    plt.scatter(meses[posicao],vendas[posicao],marker=marcador,s=tamanho,color=cor)
    plt.annotate(vendas[posicao],color=cor,xy=(meses[posicao],vendas[posicao]), xytext=(-5,5),textcoords='offset points')

#Destaque de Região
plt.axvspan(6,11,alpha=0.2,color=cores[1])
plt.text(6,80,'Recuperação de Crescimento',color='green',fontsize=9,weight='bold')

plt.box(False)
plt.tick_params(axis='x',length=0)
plt.yticks([])

plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```
