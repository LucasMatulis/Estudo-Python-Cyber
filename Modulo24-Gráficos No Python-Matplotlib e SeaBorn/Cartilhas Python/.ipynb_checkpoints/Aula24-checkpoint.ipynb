```python
#Quando só se precisa de um tipo de gráfico na figura essa interface é chamada de pyplot 
#Aqui estão algumas funções pyplot comumente usadas:
#plt.figure(): Cria uma nova figura.
#plt.plot(): Plota y versus x como linhas e/ou marcadores.
#plt.scatter(): Plota um gráfico de dispersão de y versus x.
#plt.bar(): Plota um gráfico de barras.
#plt.xlabel(): Define o rótulo do eixo x.
#plt.ylabel(): Define o rótulo do eixo y.
#plt.title(): Define um título para os eixos.
#plt.grid(): Configura as linhas da grade.
#plt.legend(): Coloca uma legenda no eixo.
#plt.show(): Exibe uma figura.
```

```python
import matplotlib.pyplot as plt
import numpy as np

meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]

#Usando o np where, crie uma lista de marcadores x para o valor mínimo e o para o resto

marcadores=np.where(vendas==np.min(vendas),'x','')

plt.plot(meses,vendas)

for posicao,marcador in enumerate(marcadores):
    if marcador=='x':
        cor='red'
        tamanho=80
    else:
        cor='blue'
        tamanho=0
    plt.scatter(meses[posicao],vendas[posicao],marker=marcador,s=tamanho,color=cor)

#Adicionando uma anotação para mostrar o menor valor em março
plt.annotate('Menor Valor',color='red',xy=(2,78),xytext=(2.2,79))

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]
marcadores=np.where(vendas==np.min(vendas),'x','')

#Paleta de Cores
cores=plt.get_cmap('Dark2').colors
ciclo_cores=cycler('color',cores)
plt.rc('axes',prop_cycle=ciclo_cores)

#Gráfico Base
plt.plot(meses,vendas)

for posicao,marcador in enumerate(marcadores):
    if marcador=='x':
        cor='red'
        tamanho=80
    else:
        cor='blue'
        tamanho=0
    plt.scatter(meses[posicao],vendas[posicao],marker=marcador,s=tamanho,color=cor)

#Adicionando uma anotação para mostrar o menor valor em março
plt.annotate('Menor Valor',color='red',xy=(2,78),xytext=(2.2,79))

#Destaque de Região
plt.axvspan(6,11,alpha=0.2,color=cores[1])
plt.text(6,80,'Recuperação de Crescimento',color='green',fontsize=9,weight='bold')

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]
marcadores=np.where(vendas==np.min(vendas),'x','')

#Paleta de Cores
cores=plt.get_cmap('Dark2').colors
ciclo_cores=cycler('color',cores)
plt.rc('axes',prop_cycle=ciclo_cores)

#Gráfico Base
plt.plot(meses,vendas)

for posicao,marcador in enumerate(marcadores):
    if marcador=='x':
        cor='red'
        tamanho=80
    else:
        cor='limegreen'
        tamanho=0
    plt.scatter(meses[posicao],vendas[posicao],marker=marcador,s=tamanho,color=cor)
    plt.annotate(vendas[posicao],color=cor,xy=(meses[posicao],vendas[posicao]), xytext=(-5,5),textcoords='offset points')

#Destaque de Região
plt.axvspan(6,11,alpha=0.2,color=cores[1])
plt.text(6,80,'Recuperação de Crescimento',color='green',fontsize=9,weight='bold')

plt.box(False)
plt.tick_params(axis='x',length=0)
plt.yticks([])

plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]


cores = plt.get_cmap('Pastel2').colors
ciclo_cores=cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.figure(figsize=(8,4))
plt.plot(meses,vendas)

#Adicionando um vspan para destacar o intervalo de meses com vendas a partir de julho
plt.axvspan(6,11,color=cores[1],alpha=0.15)

#adicionando texto para destacar o intervalo de meses com vendas a partir de Julho
plt.text(6,80,'Recuperação de crescimento', fontsize=11, color='green',weight='bold')

#usando o np where, crie uma lista de marcadores x para o valor mínimo e para o o resto
marcadores=np.where(vendas==np.min(vendas),'x','o')
for posicao,marcador in enumerate(marcadores):
    if marcador=='x':
        cor='red'
        tamanho=80
    else:
        cor=cores[0]
        tamanho=30
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador,s=tamanho, color=cor)
    plt.annotate(vendas[posicao], color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-5,5), textcoords='offset points')

plt.ylabel('Vendas (em milhares)')
plt.xlabel('Crescimento das Vendas ao Longo do Ano')

plt.yticks([])
plt.tick_params(axis='x', length=0)
plt.box(False)

plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

cores = plt.get_cmap('Pastel2').colors
ciclo_cores=cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

produtos=['Produto A','Produto B','Produto C','Produto D','Produto E']
quantidades=[23,45,17,30,37]

dados_ordenados=sorted(zip(quantidades,produtos))
quantidades_ordenadas=[x[0] for x in dados_ordenados]
produtos_ordenados=[x[1] for x in dados_ordenados]


barras=plt.barh(produtos_ordenados,quantidades_ordenadas)

plt.bar_label(barras, labels=[max(quantidades_ordenadas) if x==max(quantidades_ordenadas) else ''
                             for x in quantidades_ordenadas], padding =3, fontsize=10,fontweight='bold')


#retirando as bordas do gráfico

plt.box(False)

#retirando o eixo x
plt.xticks([])
plt.tick_params(axis='y',length=0)

#destacando a barra do produto com maior quantidade vendida

for i,barra in enumerate(barras):
    if i!= quantidades_ordenadas.index(max(quantidades_ordenadas)):
        barras[i].set_color(cores[1])
        barras[i].set_alpha(0.2)

plt.title('Quantidade de Produtos Vendidos')

#caso queira inverter o eixo, descomente a linha abaixo
#plt.gca().invert_yaxis()

plt.show()
```

## ----------------------------Matplotlib- Criando nosso mosaico------------------

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

#Ciclo de cores para os gráficos
cores=plt.get_cmap('Pastel2').colors
ciclo_cores=cycler('color',cores)
plt.rc('axes',prop_cycle=(cycler('color',plt.get_cmap('Pastel2').colors)))

#Criando o mosaico
mosaico='AA;BB;CD'
fig= plt.figure(figsize=(12,12))
espacamento={'wspace':0.3, 'hspace':0.5}# espaçamento de largura o wspace e o de altura q é o hspace para os graficos n ficarem muito juntos
axs=fig.subplot_mosaic(mosaico,gridspec_kw=espacamento)#

#Gráfico 1 - Linhas------------------------------------------------------------------------------------------------------------------

meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[85,90,78,92,88,96,94,98,100,102,105,108]

axs['A'].plot(meses,vendas)

#Adicionando um vspan para destacar o intervalo de meses com vendas a partir de julho
axs['A'].axvspan(6,11,color=cores[1],alpha=0.15)

#adicionando texto para destacar o intervalo de meses com vendas a partir de Julho
axs['A'].text(6,80,'Recuperação de crescimento', fontsize=11, color='green',weight='bold')

#usando o np where, crie uma lista de marcadores x para o valor mínimo e para o o resto
marcadores=np.where(vendas==np.min(vendas),'x','o')
for posicao,marcador in enumerate(marcadores):
    if marcador=='x':
        cor='red'
        tamanho=80
    else:
        cor=cores[0]
        tamanho=30
    axs['A'].scatter(meses[posicao], vendas[posicao], marker=marcador,s=tamanho, color=cor)
    axs['A'].annotate(vendas[posicao], color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-5,5), textcoords='offset points')

axs['A'].set_ylabel('Vendas (em milhares)')
axs['A'].set_xlabel('Crescimento das Vendas ao Longo do Ano')

axs['A'].set_yticks([])
axs['A'].tick_params(axis='x', length=0)
axs['A'].set_frame_on(False)

#Gráfico 2 - Barras-----------------------------------------------------------------------------------------------------------------------
produtos=['Produto A','Produto B','Produto C','Produto D','Produto E']
quantidades=[23,45,17,30,37]

dados_ordenados=sorted(zip(quantidades,produtos))
quantidades_ordenadas=[x[0] for x in dados_ordenados]
produtos_ordenados=[x[1] for x in dados_ordenados]

barras=axs['B'].barh(produtos_ordenados,quantidades_ordenadas)

axs['B'].bar_label(barras, labels=[max(quantidades_ordenadas) if x==max(quantidades_ordenadas) else ''
                             for x in quantidades_ordenadas], padding =3, fontsize=10,fontweight='bold')

#retirando as bordas do gráfico

axs['B'].set_frame_on(False)

#retirando o eixo x
axs['B'].set_xticks([])
axs['B'].tick_params(axis='y',length=0)

#destacando a barra do produto com maior quantidade vendida

for i,barra in enumerate(barras):
    if i!= quantidades_ordenadas.index(max(quantidades_ordenadas)):
        barras[i].set_color(cores[1])
        barras[i].set_alpha(0.2)

axs['B'].set_title('Quantidade de Produtos Vendidos')

#Gráfico 3 - pizza--------------------------------------------------------------------------------------------------------------------------

categorias=['Eletrônicos','Roupas','Alimentos','Livros']
quantidades=[45,30,15,10]
destaque=[0.1,0,0,0]#destacar a primeira fatia

axs['C'].pie(quantidades,labels=categorias, explode=destaque, autopct='%.1f')
axs['C'].set_title('Distribuição de Categorias de Produtos em Porcentagem')


#gráfico 4 -barras empilhadas--------------------------------------------------------------------------------------------------------------

anos=[2018,2019,2020,2021]
vendas_online=[5,10,15,20]
vendas_loja=[10,8,6,4]

#Usamos o plt.bar_label para colocar os valores dentros das barras e ficar mais facil o entendimento do grafico
barra_online=axs['D'].bar(anos,vendas_online,label='Online')
barra_loja=axs['D'].bar(anos,vendas_loja,bottom=vendas_online,label='Loja')
axs['D'].bar_label(barra_online,labels=vendas_online,fontsize=10,fontweight='bold',label_type='center')
axs['D'].bar_label(barra_loja,labels=vendas_online,fontsize=10,fontweight='bold',label_type='center')


#Forçar os anos a serem exibidos como inteiros
axs['D'].set_xticks(ticks=anos)
axs['D'].set_yticks([])
axs['D'].tick_params(axis='x',length=0)
axs['D'].set_frame_on(False)
axs['D'].set_ylabel('Vendas (em milhões)')
axs['D'].set_title('Vendas Online vs Loja')
axs['D'].legend()

#configurando a figura e exibindo-----------------------------------------------------------------------------------------------------------------------

#Titulo da figura

fig.suptitle('Análise de Vendas',fontsize=20,fontweight='bold',color='darkgrey')

#salvando a figura
plt.savefig('mosaico.png',dpi=300)

#mandando fazer a figura toda
plt.show()
```

## ----------------------------Matplotlib- Personalizando estilos------------------

```python
import matplotlib.pyplot as plt

produtos=['Produto A','Produto B','Produto C','Produto D','Produto E']
quantidades=[23,45,17,30,37]

plt.style.use('fivethirtyeight')

params={'axes.titlecolor':'dimgray'}#Defini o titulo com a cor cinza

plt.rcParams.update(params)

barras=plt.barh(produtos,quantidades)

plt.bar_label(barras,labels=quantidades,padding=3,fontsize=10,fontweight='bold')

plt.xticks([])
plt.title('Quantidade de Produtos Vendidos')
plt.grid(False)
plt.show()
```

```python
import matplotlib.pyplot as plt

produtos=['Produto A','Produto B','Produto C','Produto D','Produto E']
quantidades=[23,45,17,30,37]

plt.style.use('fivethirtyeight')

params={
    'axes.titlecolor':'dimgray',
    'figure.figsize':(12,6),
}

plt.rcParams.update(params)

barras=plt.barh(produtos,quantidades)

plt.bar_label(barras,labels=quantidades,padding=3,fontsize=10,fontweight='bold')

plt.xticks([])
plt.title('Quantidade de Produtos Vendidos')
plt.grid(False)
plt.show()
```
